# Futures Challenge Hedge Model

**Asset class:** Futures (ES, NQ, etc.)
**Drawdown type:** Trailing (EOD or intraday)
**Extra constraints:** Consistency rules, payout caps, minimum thresholds

## How It Works

Futures challenges differ from FX in three ways that matter for hedging:

1. **Trailing drawdown** — the blow-up level follows equity peaks, so hedge costs compound
2. **Consistency rule** — no single day’s profit can exceed X% of total, forcing multi-day hedging
3. **Payout caps** — maximum withdrawal per cycle limits how fast you recover costs

## EV Formula

$$\text{funded payout} = \min(\text{account} \times \text{target \%} \times \text{split}, \; \text{payout cap})$$

$$\text{EV} = \text{funded payout} - \text{total hedge cost} - \text{activation fee}$$

## Key Question

> "After accounting for trailing DD compounding, consistency-rule constraints, and payout caps — is there still money left?"

## 2 — Parse & Normalise Futures Fields

Futures challenges use different column names than FX. This cell normalises
the scraped data into a consistent shape for modelling.

```python
import re

def parse_dollar(val):
    """Parse '$3,000' or '3000' into float."""
    if isinstance(val, (int, float)):
        return float(val)
    if not val or val == '-' or val.lower() == 'none':
        return 0.0
    cleaned = re.sub(r'[^\d.]', '', str(val).replace(',', ''))
    return float(cleaned) if cleaned else 0.0


def parse_pct(val):
    """Parse '90%' or '0.90' or 'None' into float (0-100 scale)."""
    if isinstance(val, (int, float)):
        return float(val) if val > 1 else val * 100
    if not val or val == '-' or val.lower() == 'none':
        return 0.0
    cleaned = re.sub(r'[^\d.]', '', str(val))
    return float(cleaned) if cleaned else 0.0


def parse_consistency(val):
    """Parse consistency rule string into a ratio (0.0 = none, 0.35 = 35%)."""
    if not val or val == '-' or val.lower() == 'none':
        return 0.0
    cleaned = re.sub(r'[^\d.]', '', str(val))
    if cleaned:
        pct = float(cleaned)
        return pct / 100.0 if pct > 1 else pct
    return 0.0


def classify_dd_type(val):
    """Classify drawdown type. Futures is almost always trailing."""
    if not val:
        return "trailing"  # assume trailing for futures
    v = str(val).lower()
    if "trail" in v:
        return "trailing"
    if "static" in v or "balance" in v:
        return "static"
    return "trailing"  # default for futures


def normalise_challenge(c):
    """Normalise a raw futures challenge dict into modelling-ready shape."""
    fee = parse_dollar(c.get('fee_discounted') or c.get('fee_original') or c.get('price') or 0)
    activation = parse_dollar(c.get('activation_fee', 0))
    profit_target = parse_dollar(c.get('profit_target', 0))
    max_loss = parse_dollar(c.get('max_loss', 0))
    account_size = parse_dollar(c.get('account_size', 0))
    split = parse_pct(c.get('profit_split_pct') or c.get('profit_split', 80)) / 100.0
    payout_cap = parse_dollar(c.get('max_payout_amount', 0))
    min_threshold = parse_dollar(c.get('min_payout_threshold', 0))
    consistency_eval = parse_consistency(c.get('consistency_rule_eval', 'None'))
    consistency_funded = parse_consistency(c.get('consistency_rule_funded', 'None'))
    dd_type = classify_dd_type(c.get('max_loss_type', ''))

    # Max contracts (minis)
    minis = 0
    mc = c.get('max_contracts_minis', c.get('max_contract_size', 0))
    if isinstance(mc, str):
        nums = re.findall(r'\d+', mc)
        minis = int(nums[0]) if nums else 0
    else:
        minis = int(mc) if mc else 0

    return {
        'firm': c.get('firm', 'Unknown'),
        'account_size': account_size,
        'steps': int(c.get('steps', 1) if c.get('steps') else 1),
        'fee': fee,
        'activation_fee': activation,
        'profit_target': profit_target,
        'max_loss': max_loss,
        'dd_type': dd_type,
        'profit_split': split,
        'max_contracts_minis': minis,
        'payout_cap': payout_cap,
        'min_payout_threshold': min_threshold,
        'consistency_eval': consistency_eval,
        'consistency_funded': consistency_funded,
    }


parsed = [normalise_challenge(c) for c in challenges]
df_raw = pd.DataFrame(parsed)

print(f"Parsed {len(df_raw)} futures challenges")
print(f"\nDrawdown types: {df_raw['dd_type'].value_counts().to_dict()}")
print(f"\nConsistency rules (eval):")
print(f"  None: {(df_raw['consistency_eval'] == 0).sum()}")
print(f"  Has rule: {(df_raw['consistency_eval'] > 0).sum()}")
print(f"\nConsistency rules (funded):")
print(f"  None: {(df_raw['consistency_funded'] == 0).sum()}")
print(f"  Has rule: {(df_raw['consistency_funded'] > 0).sum()}")
print(f"\nPayout caps:")
print(f"  Uncapped: {(df_raw['payout_cap'] == 0).sum()}")
print(f"  Capped: {(df_raw['payout_cap'] > 0).sum()}")

df_raw.head(5)
```

## 3 — Futures Hedge Model: Trailing DD + Consistency Rule

### How trailing DD compounds hedge costs

- Static DD: hedge sized once, cost = (fee / DD) × target — linear.
- Trailing DD: hedge resized daily as equity rises. Each day’s loss raises the insured base,
  which raises tomorrow’s hedge size. Costs compound geometrically.

### How the consistency rule constrains the hedge

- If no single day can exceed 20% of total profit, you need at least 5 days.
- The hedge must be spread across at least `ceil(1 / threshold)` days.
- This sets a floor on the number of resizes (and therefore compounding events).

```python
def compute_futures_hedge(ch,
                         funded_target_pct=FUNDED_TARGET_PCT,
                         spread_per_contract=SPREAD_PER_CONTRACT,
                         consistency_override=None):
    """
    Futures hedge model with trailing DD compounding and consistency penalties.
    
    Returns a dict with full economics for one futures challenge.
    """
    S       = ch['account_size']
    fee     = ch['fee']
    act_fee = ch['activation_fee']
    target  = ch['profit_target']      # absolute $ target
    DD      = ch['max_loss']            # absolute $ trailing DD
    split   = ch['profit_split']
    dd_type = ch['dd_type']
    minis   = ch['max_contracts_minis'] or 1
    cap     = ch['payout_cap']
    threshold_eval    = ch['consistency_eval']
    threshold_funded  = ch['consistency_funded']
    steps   = ch['steps']

    if consistency_override is not None:
        threshold_eval = consistency_override
        threshold_funded = consistency_override

    if S <= 0 or DD <= 0 or target <= 0:
        return None  # invalid challenge

    # ── Consistency rule: minimum number of trading days ──
    if threshold_eval > 0:
        min_days = math.ceil(1.0 / threshold_eval)
    else:
        min_days = 1

    # N = number of daily hedge events (at least min_days)
    N = max(min_days, 10)  # at least 10 resizes for realism

    # ── Walk the eval phase(s) ──
    L = fee  # insured base starts at the challenge fee
    phases = []

    for step in range(steps):
        step_target = target / steps  # split target across steps
        daily_move = step_target / N
        L_start = L
        total_spread = 0.0

        if dd_type == 'trailing':
            # Trailing: each day, hedge = L / DD. Prop side gains daily_move,
            # so hedge side loses proportionally. DD trails up.
            for day in range(N):
                hedge_size = L / DD  # notional hedge sized to recover L on DD event
                hedge_loss = hedge_size * (daily_move / S) * S  # simplified: daily_move / DD * L ... let's be precise
                # Actually: prop gains daily_move on account S.
                # Hedge is short hedge_size notional. It loses (daily_move / S) * hedge_size * S
                # More precisely: hedge loses daily_move * (hedge_size / S) * S  ... no.
                # The hedge is sized so that a DD-dollar drop recovers L.
                # hedge_size / S * DD = L  =>  hedge_size = L * S / DD
                # When prop goes UP by daily_move, hedge loses: (daily_move / S) * hedge_size = daily_move * L / DD
                hedge_loss = daily_move * L / DD
                day_spread = minis * spread_per_contract * 2  # round-trip
                total_spread += day_spread
                L += hedge_loss + day_spread
        else:
            # Static: single calc
            hedge_size = L * S / DD if DD > 0 else 0
            hedge_loss = target * L / DD
            total_spread = minis * spread_per_contract * 2 * N
            L += hedge_loss + total_spread

        phase_cost = L - L_start
        phases.append({
            'phase': step + 1,
            'target': round(step_target, 2),
            'days': N,
            'insured_start': round(L_start, 2),
            'insured_end': round(L, 2),
            'phase_cost': round(phase_cost, 2),
            'spread_cost': round(total_spread, 2),
        })

    total_cost = L + act_fee

    # ── Funded payout ──
    raw_payout = S * funded_target_pct * split
    if cap > 0:
        funded_payout = min(raw_payout, cap)
    else:
        funded_payout = raw_payout

    EV = funded_payout - total_cost
    breakeven_payout = total_cost / split if split > 0 else 0
    breakeven_pct = (breakeven_payout / S) * 100 if S > 0 else 0

    # ── Payout delay (cycles to recover total cost) ──
    if cap > 0 and funded_payout > 0:
        cycles_to_recover = math.ceil(total_cost / (cap * split)) if cap * split > 0 else 999
    else:
        cycles_to_recover = 1

    # ── Capital requirement ──
    margin_per_mini = 1000  # approximate ES margin
    capital_required = minis * margin_per_mini * 2  # margin + buffer

    return {
        'firm': ch['firm'],
        'account_size': S,
        'fee': round(fee, 2),
        'activation_fee': round(act_fee, 2),
        'steps': steps,
        'profit_target': round(target, 2),
        'max_loss': round(DD, 2),
        'dd_type': dd_type,
        'profit_split_pct': round(split * 100, 1),
        'max_contracts': minis,
        'payout_cap': round(cap, 2),
        'consistency_eval': round(threshold_eval, 2),
        'consistency_funded': round(threshold_funded, 2),
        'min_trading_days': min_days,
        'phases': phases,
        'total_cost': round(total_cost, 2),
        'funded_payout': round(funded_payout, 2),
        'EV': round(EV, 2),
        'breakeven_payout': round(breakeven_payout, 2),
        'breakeven_pct': round(breakeven_pct, 2),
        'cycles_to_recover': cycles_to_recover,
        'capital_required': round(capital_required, 2),
        'capital_efficiency': round(EV / capital_required, 4) if capital_required > 0 else 0,
    }


# Run on all parsed challenges
results = [compute_futures_hedge(ch) for ch in parsed]
results = [r for r in results if r is not None]  # drop invalid
df = pd.DataFrame(results)

pos = (df['EV'] > 0).sum()
neg = (df['EV'] <= 0).sum()
print(f"\u2705 Positive EV: {pos}/{len(df)} challenges")
print(f"\u274c Negative EV: {neg}/{len(df)} challenges")
print(f"\ud83d\udcca Average EV: ${df['EV'].mean():,.2f}")
if pos > 0:
    print(f"\ud83c\udfaf Best EV: {df.loc[df['EV'].idxmax(), 'firm']} \u2192 ${df['EV'].max():,.2f}")
print(f"\n\u2500\u2500 EV by Drawdown Type \u2500\u2500")
for cat in ['trailing', 'static']:
    mask = df['dd_type'] == cat
    if mask.any():
        print(f"  {cat:10s} ({mask.sum()} challenges): avg EV = ${df.loc[mask, 'EV'].mean():,.2f}")
```

## 6 — Trailing DD Compounding: Cost Stack Growth

How the insured base (cost stack) grows as the prop account profits day by day.
Each day’s hedge loss raises the insured base, which raises the next day’s hedge size.

```python
# Pick top 5 by EV and trace their daily insured-base growth
top5 = df.nlargest(5, 'EV')
top5_keys = set(zip(top5['firm'], top5['account_size']))
top5_parsed = [ch for ch in parsed if (ch['firm'], ch['account_size']) in top5_keys]

fig, ax = plt.subplots(figsize=(12, 6))
colors = plt.cm.tab10.colors

for i, ch in enumerate(top5_parsed):
    S = ch['account_size']
    DD = ch['max_loss']
    target = ch['profit_target']
    fee = ch['fee']
    minis = ch['max_contracts_minis'] or 1
    threshold = ch['consistency_eval']
    min_days = math.ceil(1.0 / threshold) if threshold > 0 else 1
    N = max(min_days, 10)
    daily_move = target / N

    L = fee
    trace = [L]
    for day in range(N):
        hedge_loss = daily_move * L / DD
        spread = minis * SPREAD_PER_CONTRACT * 2
        L += hedge_loss + spread
        trace.append(L)

    label = f"{ch['firm']} (${ch['account_size']/1000:.0f}k)"
    ax.plot(range(N + 1), trace, marker='o', markersize=3, linewidth=2.5,
            label=label, color=colors[i % 10])

ax.set_xlabel('Trading Day')
ax.set_ylabel('Insured Base / Cost Stack ($)')
ax.set_title('Trailing DD Compounding \u2014 Daily Cost Stack Growth (Top 5)')
ax.legend(fontsize=9, loc='upper left')
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7 — Consistency Rule Impact

What happens to EV as the consistency threshold changes from 10% (strict) to 100% (no rule)?

```python
top5_idx = df.nlargest(5, 'EV').index.tolist()
top5_ch = [parsed[i] for i in top5_idx]

thresholds = [0.10, 0.15, 0.20, 0.25, 0.30, 0.40, 0.50, 1.00]

fig, ax = plt.subplots(figsize=(12, 6))
colors = plt.cm.tab10.colors

for i, ch in enumerate(top5_ch):
    evs = []
    for t in thresholds:
        r = compute_futures_hedge(ch, consistency_override=t)
        evs.append(r['EV'] if r else 0)
    label = f"{ch['firm']} (${ch['account_size']/1000:.0f}k)"
    ax.plot([t * 100 for t in thresholds], evs, marker='o', markersize=6,
            linewidth=2.5, label=label, color=colors[i % 10])

ax.axhline(y=0, color='black', linewidth=1, linestyle='--')
ax.axvline(x=CONSISTENCY_THRESHOLD * 100, color='#3498db', linewidth=2.5,
           linestyle='--', alpha=0.8, label=f'Default: {CONSISTENCY_THRESHOLD*100:.0f}%')

ax.set_xlabel('Consistency Threshold (% of total profit per day)')
ax.set_ylabel('EV ($)')
ax.set_title('EV Sensitivity to Consistency Rule Strictness\n'
             '(Lower % = stricter rule = more trading days = more compounding)')
ax.legend(fontsize=9, loc='lower right')
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 12 — Summary

### What this notebook tells you

- Futures challenges are structurally harder to hedge than FX due to trailing drawdown compounding.
- The consistency rule forces multi-day hedging, adding compounding events.
- Payout caps extend recovery time and reduce capital efficiency.
- Despite this, some futures challenges still produce positive EV under the insurance model.

### Key differences from FX notebooks

| Factor | FX (Type A/B/C) | Futures |
|--------|-----------------|----------|
| Drawdown | Usually static | Usually trailing |
| Consistency rule | Rare | Common (20–50%) |
| Payout caps | Uncommon | Very common ($1.5K–$10K) |
| Min payout threshold | Rare | Common ($250–$3K) |
| Activation fee | None | Some firms charge |
| Funded target | ~8% | ~6% (lower) |
| Hedge cost growth | Linear (static DD) | Geometric (trailing DD) |

### When futures hedging makes sense

- Larger account sizes ($100K+) where the payout exceeds compounding costs
- Low-fee firms with generous splits (90%+)
- Firms with no or lenient consistency rules
- Firms with high or no payout caps

### When to avoid

- Small accounts ($10K–$25K) where fee + compounding eats the payout
- Strict consistency (10–15%) with tight trailing DD
- Firms with low payout caps ($500–$1K) that make recovery take too many cycles
